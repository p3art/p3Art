```python
# 程式碼A區==程式碼A區==程式碼A區==程式碼A區==程式碼A區
# AAAAAAAA==AAAAAAAA==AAAAAAAAA==AAAAAAAA==AAAAAAAA
#
# Sine-Cosine-2
# 2021-09-21
#
# python 3.8
# run on https://replit.com
# run on ubuntu 16.04 LTS
#
# 要執行本程式, 請按上方run▶
#
# Browser encoding UTF-8 HTML5 URL / as %2F
#
# 程式碼B區==程式碼B區==程式碼B區==程式碼B區==程式碼B區
# BBBBBBBB==BBBBBBBB==BBBBBBBBB==BBBBBBBB==BBBBBBBB
#
# to fix matplotlib warning
import os
os.environ['MPLCONFIGDIR'] = os.getcwd() + "/configs/"
#
import sys
#
# 載入繪圖模組 pyplot, 重新命名為 plot
import matplotlib.pyplot as plot
import numpy as np
from shapely.geometry  import LineString
#
# 定義圖面函數 ------------------------------------------------------
def set_graphic_area(width,height) :
    #    
    cm2inch = 1/2.54    # inch per cm
    #
    # define graphic area
    #
    left_margin  = 2.0  # cm
    right_margin = 1.0  # cm
    #
    figure_width  = width  # cm , from xmin to xmax
    figure_height = height # cm , from ymin to ymax
    #
    top_margin = 1.0    # cm
    bottom_margin = 1.0 # cm
    #
    box_width = left_margin + figure_width + right_margin   # cm
    box_height = top_margin + figure_height + bottom_margin # cm
    #
    top_value    = 1.0 - top_margin / box_height
    bottom_value = bottom_margin / box_height
    left_value   = left_margin / box_width
    right_value  = 1.0 - right_margin / box_width
    #
    return (box_width*cm2inch,box_height*cm2inch,top_value,bottom_value,left_value,right_value,width)
#
# end of def
#
# Define the arc
# center position is cxy = (cx, cy)
# start_angle, end_angle is in degrees
#
def shapely_Arc(cxy, r, start_angle, end_angle) :
    #
    numsegments = 1440
    # The coordinates of the arc
    theta = np.radians(np.linspace(start_angle, end_angle, numsegments))
    x = cxy[0] + r * np.cos(theta)
    y = cxy[1] + r * np.sin(theta)
    Arc = LineString(np.column_stack([x, y]))
    return Arc
    #
# end of def
#
def motif_line_1(LS, COLOR, ZORDER, LINEWIDTH) :
    #
    xs, ys = LS.xy
    ax_1.plot(xs, ys, color=COLOR, alpha=0.7, linewidth=LINEWIDTH, solid_capstyle='round', zorder=ZORDER)
    return
#
def motif_line_2(LS, COLOR, ZORDER, LINEWIDTH) :
    #
    xs, ys = LS.xy
    ax_2.plot(xs, ys, color=COLOR, alpha=0.7, linewidth=LINEWIDTH, solid_capstyle='round', zorder=ZORDER)
    return
#
def motif_fill(LS, COLOR, ZORDER, ALPHA) :
    #
    xs, ys = LS.xy
    ax_1.fill(xs, ys, alpha=ALPHA, closed=True, edgecolor=None, facecolor=COLOR, fill=True, joinstyle='miter', zorder=ZORDER)
    return
#
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)
    # end if
# end def
#
# 程式碼C區==程式碼C區==程式碼C區==程式碼C區==程式碼C區
# CCCCCCCC==CCCCCCCC==CCCCCCCCC==CCCCCCCC==CCCCCCCC
#
# 設定圖面大小
#  setup圖面大小 (10,5) 單位公分
#  tup7 is type of Tuple, have 7 indexed value
#
# main()
tup7 = set_graphic_area(4*np.pi,8)  
# 呼叫 pyplot.figure(), 建立一個圖表物件, 並成為目前圖表物件
fig = plot.figure(figsize=(tup7[0], tup7[1])) 
# 圖表的繪圖區域被分為1個子圖, 1 row, 1 column, plot to index 1
ax_1 = fig.add_subplot(2,1,1)   
fig.subplots_adjust(
                top    = tup7[2] ,
                bottom = tup7[3] ,
                left   = tup7[4] ,
                right  = tup7[5] ,
                )
#
plot.xlim(-2*np.pi, 2*np.pi)    # 設定X軸的顯示範圍, from xmin to xmax
plot.ylim(-1.5, 1.5)            # 設定Y軸的顯示範圍, from ymin to ymax
#
xvalues = [-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0,
            np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
#
yvalues = [-1.5,-1,-0.5,0,0.5,1,1.5]
#
plot.xticks(xvalues)
plot.yticks(yvalues)
plot.tick_params(labelsize=6)
#
ax_1.xaxis.set_major_formatter(plot.FuncFormatter(format_func))
#
plot.xlabel("x")
plot.ylabel("sin(x)")
#
# get current axes, set X,Y same ratio & scale 
# 'equal' --> 0.5
plot.gca().set_aspect('equal', adjustable='box')
#
plot.grid(color='green', linewidth=0.25)
ax_1.grid(True)
# ax.grid(False)
ax_1.set_axis_on() # will turn on  grid line
# ax.set_axis_off()  # will turn off grid line
#
H_line = LineString([(-2*np.pi, 0), (2*np.pi, 0)])
motif_line_1(H_line, 'red', 10, 0.4)
#
plot.savefig("./png/2-1.png", format="png", dpi=150)
#
# 程式碼D區==程式碼D區==程式碼D區==程式碼D區==程式碼D區
# DDDDDDDD==DDDDDDDD==DDDDDDDDD==DDDDDDDD==DDDDDDDD
#
numsegments = 1440
# The coordinates of the Sine
x = np.radians(np.linspace(-360, 360, numsegments))
y = np.sin(x)
Sine = LineString(np.column_stack([x, y]))
motif_line_1(Sine, 'blue', 10, 1.0)
#
plot.savefig("./png/2-2.png", format="png", dpi=150)
#
ax_2 = fig.add_subplot(2,1,2)   
fig.subplots_adjust(
                top    = tup7[2] ,
                bottom = tup7[3] ,
                left   = tup7[4] ,
                right  = tup7[5] ,
                )
#
plot.xlim(-2*np.pi, 2*np.pi)    # 設定X軸的顯示範圍, from xmin to xmax
plot.ylim(-1.5, 1.5)            # 設定Y軸的顯示範圍, from ymin to ymax
#
xvalues = [-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0,
            np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
#
yvalues = [-1.5,-1,-0.5,0,0.5,1,1.5]
#
plot.xticks(xvalues)
plot.yticks(yvalues)
plot.tick_params(labelsize=6)
#
ax_2.xaxis.set_major_formatter(plot.FuncFormatter(format_func))
#
plot.xlabel("x")
plot.ylabel("cos(x)")
#
# get current axes, set X,Y same ratio & scale 
plot.gca().set_aspect('equal', adjustable='box')
#
plot.grid(color='green', linewidth=0.25)
ax_2.grid(True)
# ax.grid(False)
ax_2.set_axis_on() # will turn on  grid line
# ax.set_axis_off()  # will turn off grid line
#
motif_line_2(H_line, 'red', 10, 0.4)
#
plot.savefig("./png/2-3.png", format="png", dpi=150)
#
numsegments = 1440
# The coordinates of the Sine
x = np.radians(np.linspace(-360, 360, numsegments))
y = np.cos(x)
Sine = LineString(np.column_stack([x, y]))
motif_line_2(Sine, 'magenta', 10, 1.0)
#
plot.savefig("./png/2-4.png", format="png", dpi=150)
#
try:
    print('Press Ctrl-C to exit')
    plot.show()
except KeyboardInterrupt:
    # press Ctrl-C to quit
    os.system('clear')
    sys.exit()
```
