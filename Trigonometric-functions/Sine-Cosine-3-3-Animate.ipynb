```python
# 程式碼A區==程式碼A區==程式碼A區==程式碼A區==程式碼A區
# AAAAAAAA==AAAAAAAA==AAAAAAAAA==AAAAAAAA==AAAAAAAA
#
# Sine-Cosine-3-Animate-3
# 2021-09-21
#
# python 3.8
# run on https://replit.com
# run on ubuntu 16.04 LTS
#
# 要執行本程式, 請按上方run▶
#
# Browser encoding UTF-8 HTML5 URL / as %2F
#
# 程式碼B區==程式碼B區==程式碼B區==程式碼B區==程式碼B區
# BBBBBBBB==BBBBBBBB==BBBBBBBBB==BBBBBBBB==BBBBBBBB
#
# Enable interactive plot
%matplotlib widget
#
# to fix matplotlib warning
import os
os.environ['MPLCONFIGDIR'] = os.getcwd() + "/configs/"
#
import sys
import IPython
import numpy as np
import datetime as dt
#
# 載入繪圖模組 pyplot, 重新命名為 plt
import matplotlib.pyplot as plt
from shapely.geometry  import LineString
from matplotlib.animation import FuncAnimation
#
# 定義圖面函數 ------------------------------------------------------
def set_graphic_area(width,height) :
    #    
    cm2inch = 1/2.54    # inch per cm
    #
    # define graphic area
    #
    left_margin  = 2.0  # cm
    right_margin = 1.0  # cm
    #
    figure_width  = width  # cm , from xmin to xmax
    figure_height = height # cm , from ymin to ymax
    #
    top_margin = 1.0    # cm
    bottom_margin = 1.0 # cm
    #
    box_width = left_margin + figure_width + right_margin   # cm
    box_height = top_margin + figure_height + bottom_margin # cm
    #
    top_value    = 1.0 - top_margin / box_height
    bottom_value = bottom_margin / box_height
    left_value   = left_margin / box_width
    right_value  = 1.0 - right_margin / box_width
    #
    return (box_width*cm2inch,box_height*cm2inch,top_value,bottom_value,left_value,right_value,width)
#
# end of def
#
def motif_line(LS, COLOR, ZORDER, LINEWIDTH) :
    #
    xs, ys = LS.xy
    ax.plot(xs, ys, color=COLOR, alpha=0.7, linewidth=LINEWIDTH, solid_capstyle='round', zorder=ZORDER)
    return
#
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)
    # end if
# end def
#
# 程式碼C區==程式碼C區==程式碼C區==程式碼C區==程式碼C區
# CCCCCCCC==CCCCCCCC==CCCCCCCCC==CCCCCCCC==CCCCCCCC
#
# 設定圖面大小
#  setup圖面大小 (4*np.pi,3) 單位公分
#  tup7 is type of Tuple, have 7 indexed value
#
# main()
tup7 = set_graphic_area(4*np.pi,3)  
# 呼叫 pyplot.figure(), 建立一個圖表物件, 並成為目前圖表物件
fig = plt.figure(figsize=(tup7[0], tup7[1])) 
# 圖表的繪圖區域被分為1個子圖, 1 row, 1 column, plot to index 1
ax = fig.add_subplot(1,1,1)   
fig.subplots_adjust(
                top    = tup7[2] ,
                bottom = tup7[3] ,
                left   = tup7[4] ,
                right  = tup7[5] ,
                )
#
plt.xlim(-2*np.pi, 2*np.pi)    # 設定X軸的顯示範圍, from xmin to xmax
plt.ylim(-1.5, 1.5)            # 設定Y軸的顯示範圍, from ymin to ymax
#
xvalues = [-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0,
            np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
#
yvalues = [-1.5,-1,-0.5,0,0.5,1,1.5]
plt.xticks(xvalues)
plt.yticks(yvalues)
plt.tick_params(labelsize=6)
#
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
#
plt.title('Sine & Cosine Curve')
plt.xlabel("x")
plt.ylabel("sin(x) cos(x)")
#
# get current axes, set X,Y same ratio & scale 
plt.gca().set_aspect('equal', adjustable='box')
#
plt.grid(color='green', linewidth=0.25)
ax.grid(True)
# ax.grid(False)
ax.set_axis_on() # will turn on  grid line
# ax.set_axis_off()  # will turn off grid line
#
H_line = LineString([(-2*np.pi, 0), (2*np.pi, 0)])
motif_line(H_line, 'darkviolet', 15, 0.5)
#
plt.savefig("./png/3-1.png", format="png", dpi=150)
#
# 程式碼D區==程式碼D區==程式碼D區==程式碼D區==程式碼D區
# DDDDDDDD==DDDDDDDD==DDDDDDDDD==DDDDDDDD==DDDDDDDD
#
# each segment 10 degree
#
numsegments = 73
x_coords_degree = np.linspace(-360, 360, numsegments)
x_coords_radian = np.radians(x_coords_degree)
# print(len(x_coords_degree))
# print(x_coords_degree)
#
# The coordinates of the Sine & Cosine
xs = x_coords_radian
sin_ys = np.sin(xs)
cos_ys = np.cos(xs)
#
# function that draws each frame of the animation
def animate(i, X_list, Sy_list, Cy_list):
    #
    if i > 73 :
        print('please restart kernel')
        #automatically restarts kernel
        IPython.Application.instance().kernel.do_shutdown(True)
    # end if
    #
    ax.clear()
    #
    plt.xlim(-2*np.pi, 2*np.pi)    # 設定X軸的顯示範圍, from xmin to xmax
    plt.ylim(-1.5, 1.5)            # 設定Y軸的顯示範圍, from ymin to ymax
    #
    xvalues = [-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0,
            np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
    #
    yvalues = [-1.5,-1,-0.5,0,0.5,1,1.5]
    plt.xticks(xvalues)
    plt.yticks(yvalues)
    plt.tick_params(labelsize=6)
    ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
    plt.grid(color='green', linewidth=0.25)
    #
    plt.title('Sine & Cosine Curve')
    plt.xlabel("x")
    plt.ylabel("sin(x) cos(x)")
    #
    plt.subplots_adjust(bottom=0.20)
    #    
    X_list.append(xs[i])
    Sy_list.append(sin_ys[i])
    Cy_list.append(cos_ys[i])
    #
    ax.plot(X_list, Sy_list, color='blue')
    ax.plot(X_list, Cy_list, color='red')
    motif_line(H_line, 'darkviolet', 15, 0.5)
    #    
# end def
#
# empty lists for the x,y data
Xs = []
SIN_ys = []
COS_ys = []
#
# run the animation
i = 0
ani = FuncAnimation(fig, animate, fargs=(Xs,SIN_ys,COS_ys), interval=200)
#
```
