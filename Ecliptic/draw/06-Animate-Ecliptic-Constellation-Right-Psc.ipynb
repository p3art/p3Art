```python
# 2021-09-26
# 06-Animate-Ecliptic-Constellation-Right
#
# When into animaton mode, code error massage will not display, include pint()
#
# How to guide ...
# https://holypython.com/how-to-save-matplotlib-animations-the-ultimate-guide/
#
# Cell Markdown
# ![天球的赤道座標](./z-天球的赤道座標.png)
#
# Enable interactive plot
%matplotlib widget
#
import os
path = os.getcwd()
print('當前工作目錄(Current working directory) ==>', path)
#
import csv
import math
import numpy  as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fontManager
#
from matplotlib import animation
from IPython.display  import HTML
from shapely.geometry import LineString
#
# import matplotlib as mpl 
# mpl.rcParams['animation.ffmpeg_path'] = r'D:/0_sdr/ffmpeg-4.4-essentials_build/bin/ffmpeg.exe'
# 
# 定義圖面函數 ------------------------------------------------------
def set_graphic_area(width,height) :
    #    
    cm2inch = 1/2.54    # inch per cm
    #
    # define graphic area
    #
    left_margin  = 2.0  # cm
    right_margin = 1.0  # cm
    #
    figure_width  = width  # cm , from xmin to xmax
    figure_height = height # cm , from ymin to ymax
    #
    top_margin = 1.0    # cm
    bottom_margin = 1.5 # cm
    #
    box_width = left_margin + figure_width + right_margin   # cm
    box_height = top_margin + figure_height + bottom_margin # cm
    #
    top_value    = 1.0 - top_margin / box_height
    bottom_value = bottom_margin / box_height
    left_value   = left_margin / box_width
    right_value  = 1.0 - right_margin / box_width
    #
    return (box_width*cm2inch,box_height*cm2inch,top_value,bottom_value,left_value,right_value,width)
#
# end of def
#
# Define the arc
# center position is cxy = (cx, cy)
# start_angle, end_angle is in degrees
#
def shapely_Arc(cxy, r, start_angle, end_angle) :
    #
    numsegments = 1440
    # The coordinates of the arc
    theta = np.radians(np.linspace(start_angle, end_angle, numsegments))
    x = cxy[0] + r * np.cos(theta)
    y = cxy[1] + r * np.sin(theta)
    Arc = LineString(np.column_stack([x, y]))
    return Arc
    #
# end of def
#
def motif_line(LS, COLOR, ZORDER, LINEWIDTH) :
    #
    xs, ys = LS.xy
    ax.plot(xs, ys, color=COLOR, alpha=0.7, linewidth=LINEWIDTH, solid_capstyle='round', zorder=ZORDER)
    return
#
def motif_fill(LS, color, ZORDER, Alpha) :
    #
    xs, ys = LS.xy
    ax.fill(xs, ys, alpha=Alpha, closed=True, edgecolor=None, facecolor=color, fill=True, zorder=ZORDER)
    return
#
# find number of multiples of 60° (π/6)
#
def format_func(value, tick_number):
    N = int(np.round(6 * value / np.pi))
    return '$' + str(-2*N) + '^h$'
# end def
#
# 設定圖面大小
#  tup7 is type of Tuple, have 7 indexed value
#
tup7 = set_graphic_area(9*np.pi,9)  
# 呼叫 pyplot.figure(), 建立一個圖表物件, 並成為目前圖表物件
fig = plt.figure(figsize=(tup7[0], tup7[1])) 
# 圖表的繪圖區域被分為1個子圖, 1 row, 1 column, plot to index 1
ax = fig.add_subplot(1,1,1)   
fig.subplots_adjust(
                top    = tup7[2] ,
                bottom = tup7[3] ,
                left   = tup7[4] ,
                right  = tup7[5] ,
                )
#
plt.xlim(-2*np.pi, np.pi)    # 設定X軸的顯示範圍, from xmin to xmax
plt.ylim(-1.5, 1.5)          # 設定Y軸的顯示範圍, from ymin to ymax
#
xvalues = [-12*np.pi/6, -11*np.pi/6, -10*np.pi/6, -9*np.pi/6, -8*np.pi/6,
            -7*np.pi/6,  -6*np.pi/6,  -5*np.pi/6, -4*np.pi/6, -3*np.pi/6, 
            -2*np.pi/6,  -1*np.pi/6, 0]
#
yvalues = [-1.5,-1,-0.5,0,0.5,1,1.5]
#
plt.xticks(xvalues)
plt.yticks(yvalues)
plt.tick_params(labelsize=11)
#
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
#
plt.xlabel("RA Hours",fontsize=12)
plt.ylabel("Sin(Dec)",fontsize=12)
#
# get current axes, set X,Y same ratio & scale 
# 'equal' --> 0.5
plt.gca().set_aspect('equal', adjustable='box')
#
plt.grid(color='greenyellow', linewidth=0.5)
ax.grid(True)
# ax.grid(False)
ax.set_axis_on() # will turn on  grid line
# ax.set_axis_off()  # will turn off grid line
#
circle_1 = shapely_Arc((np.pi/2,0), 1, 0, 360)
motif_line(circle_1, 'blue', 10, 0.8)
#
RA_line = LineString([(-2*np.pi,0), (0,0)])
motif_line(RA_line, 'red', 30, 0.8)
#
degree = np.pi/180
ya = np.sin(23.5*degree)
#
Ecliptic_max = LineString([(-2*np.pi, ya), (0, ya)])
Ecliptic_min = LineString([(-2*np.pi,-ya), (0,-ya)])
motif_line(Ecliptic_max, 'darkviolet', 30, 0.4)
motif_line(Ecliptic_min, 'darkviolet', 30, 0.4)
#
myfont = fontManager.FontProperties(fname='fonts/msjh.ttc')
#
text_1_xy = (0+0.05, ya)
text_2_xy = (0+0.025,-ya-0.1)
ax.annotate("23.5°", text_1_xy,
            color='darkviolet',
            fontproperties=myfont,
            fontsize=9, 
            zorder=10)
#
ax.annotate("-23.5°", text_2_xy,
            color='darkviolet',
            fontproperties=myfont,
            fontsize=9, 
            zorder=10)
#
plt.savefig("./graphic/06-1.png", format="png", dpi=150)
#
## -----------------------------------
#
ConName1_dict = {} #使用大刮號{}創建字典
ConName2_dict = {}
#
def Con_dict_add(data1, data2, data3) :
    global ConName1_dict, ConName2_dict

    ConName1_dict[data1] = data2
    ConName2_dict[data1] = data3
    #
    return               
# end of def
#
csvf3 = open('Constellations_Name.csv', newline='', encoding="utf-8")
# csvf3檔儲存赤道天區各星座之 [編號, 西洋簡稱, 中文名稱]
# ex: 01,And,仙女座
#     02,Ant,唧筒座
# 逐筆讀入資料, 並存於 dictionary
for row in csv.reader(csvf3, delimiter=','):
    Con_dict_add(row[0], row[1], row[2])
csvf3.close()
#
## -----------------------------------
# 黃道十三星座代碼表
# ------------------------------------
# Psc雙魚座66(跨越天球本初子午線)
# Ari白羊座07、Tau金牛座78、Gem雙子座38、
# Cnc巨蟹座12、Leo獅子座46、Vir室女座86、
# Lib天秤座49、Sco天蠍座73、Oph蛇夫座59、
# Sgr人馬座72、Cap摩羯座16、Aqr寶瓶座04。
# ------------------------------------
# 04, 5, Ecliptic, Aqr, 寶瓶座
# 07, 5, Ecliptic, Ari, 白羊座
# 12, 2, Ecliptic, Cnc, 巨蟹座
# 16, 5, Ecliptic, Cap, 摩羯座
# 38, 5, Ecliptic, Gem, 雙子座
# 46, 4, Ecliptic, Leo, 獅子座
# 49, 2, Ecliptic, Lib, 天秤座
# 59, 5, Ecliptic, Oph, 蛇夫座
# 66, 3, Ecliptic, Psc, 雙魚座, cross0H ; 66_bound.dat is actually 66_bound_new.dat
# 72, 5, Ecliptic, Sgr, 人馬座
# 73, 2, Ecliptic, Sco, 天蠍座
# 78, 3, Ecliptic, Tau, 金牛座
# 86, 4, Ecliptic, Vir, 室女座
#
path_data    = './data/'
path_bound   = './bound/'
path_merge   = './merge/'
path_graphic = './graphic/'
#
def read_boundary_data(fileHead) :
    # 逐筆讀入(星座邊界)資料
    # o Type of point --> Original
    # I Type of point --> Interpolated(內插)
    #
    global bound_xs, bound_ys, f6ListX, f6ListY
    csvf6 = open(path_bound + fileHead + '_bound.dat', newline='') 
    for row in csv.reader(csvf6, delimiter=','):
        if eval(row[0]) == 1001 :
            f6ListX.append(bound_xs)
            f6ListY.append(bound_ys)
            bound_xs = []
            bound_ys = []
        else :
            my_radian = eval(row[0])*15*degree
            bound_xs.append(-my_radian)
            bound_ys.append(np.sin(eval(row[1])*degree))
        #end if
    # end for    
    csvf6.close()
# end def
#
total_xs = []
total_ys = []
#
def do_A_code_block() :
    global total_xs, total_ys
    xs = f6ListX[0]
    ys = f6ListY[0]
    ax.plot(xs, ys, color='royalblue', alpha=1.0, linewidth=0.7, linestyle='dotted', solid_capstyle='round', zorder=20)
    ax.fill(xs, ys, alpha=1.0, closed=True, edgecolor='none', facecolor=[228/255,235/255,243/255], fill=True, zorder=10)
    total_xs.append(xs)
    total_ys.append(ys)
# end def
#
def do_B_code_block() :
    xyList = []
    for i in range(len(f6ListX)) :
        xyList.append(zip(f6ListX[i], f6ListY[i]))
    # end for
    Line_num = len(xyList)
    print ('Line_num is',Line_num)
    print ('\n')    
    for j, data in enumerate(xyList) :
        # test block
        if j == 0 :
            print ('line 1 is red')
            Line1 = LineString(data)
            xs, ys = Line1.xy
            xs_list = []
            for x in xs :
                xs_list.append(-x)
            # end for
            ys_list = []
            for y in ys :
                ys_list.append(y)
            # end for            
            Line1_xs =     [0] + xs_list + [0]
            Line1_ys = [ys[0]] + ys_list + [ys[-1]]
            print ('Line1 first xy', (Line1_xs[ 0],Line1_ys[ 0]))
            print ('Line1  last xy', (Line1_xs[-1],Line1_ys[-1]))
            print ('\n')       
            # 
            #ax.plot(Line1_xs, Line1_ys, color='red', dashes=[6, 2], alpha=1.0, linewidth=1.0, solid_capstyle='round', zorder=200)
            #plt.savefig('./graphic/' + '06-step-3A.png', dpi=150)
        if j == 1 :
            print ('line 2 is blue')
            Line2 = LineString(data) 
            xs, ys = Line2.xy
            xs_list = []
            for x in xs :
                xs_list.append(-x-2*np.pi)
            # end for
            ys_list = []
            for y in ys :
                ys_list.append(y)
            # end for            
            Line2_xs = [-2*np.pi] + xs_list + [-2*np.pi]
            Line2_ys = [ys[0]]    + ys_list + [ys[-1]]
            print ('Line2 first xy', (Line2_xs[ 0],Line2_ys[ 0]))
            print ('Line2  last xy', (Line2_xs[-1],Line2_ys[-1]))
            print ('\n')
            #
            #ax.plot(Line2_xs, Line2_ys, color='blue', dashes=[6, 2], alpha=1.0, linewidth=1.0, solid_capstyle='round', zorder=200)
            #plt.savefig('./graphic/' + '06-step-3B.png', dpi=150)
    # end for
    #
    if Line_num == 2 :
        total_xs.append(Line1_xs)
        total_ys.append(Line1_ys)
        total_xs.append(Line2_xs)
        total_ys.append(Line2_ys)
        #
        ax.plot(Line1_xs, Line1_ys, color='royalblue', alpha=1.0, linewidth=0.7, linestyle='dotted', solid_capstyle='round', zorder=20)
        ax.plot(Line2_xs, Line2_ys, color='royalblue', alpha=1.0, linewidth=0.7, linestyle='dotted', solid_capstyle='round', zorder=20)
        ax.fill(Line1_xs, Line1_ys, alpha=1.0, closed=True, edgecolor='none', facecolor=[228/255,235/255,243/255], fill=True, zorder=10)        
        ax.fill(Line2_xs, Line2_ys, alpha=1.0, closed=True, edgecolor='none', facecolor=[228/255,235/255,243/255], fill=True, zorder=10)        
    # end if
# end def
#
## ===================================================================================================================
#
### 繪製星座邊界, 星座邊界檔可於IAU網頁下載
#
key_set = [('04', 5) , ('07', 5) , ('12', 2) , ('16', 5) , 
           ('38', 5) , ('46', 4) , ('49', 2) , ('59', 5) , ('66', 3) ,
           ('72', 5) , ('73', 2) , ('78', 3) , ('86', 4) ]
#
for kset in key_set :
    fileHead = kset[0]
    key = fileHead
    k = kset[1]        # means has k star asterism lines (星群連線)
    #
    print (key, ConName1_dict[key].upper(), ConName2_dict[key])
    #
    bound_xs = []
    bound_ys = []
    f6ListX = []
    f6ListY = []
    #
    read_boundary_data(fileHead)
    #
    # for GEM
    # f6ListX is [[x01,x02,x03, ..]]
    # f6ListY is [[y01,y02,y03, ..]]
    #
    if len(f6ListX) == 1 :
        # 這是簡單的封閉區域 
        do_A_code_block()
    else : 
        do_B_code_block()
    # end if
    #
    # print ('\n')
    # pprint.pprint (f6ListX)
    # print('星座邊界: X座標區塊有',len(f6ListX),'組資料')
    # print('\n')
    #
# end for
#
plt.savefig("./graphic/0-2.png", format="png", dpi=150)
#
## -----------------------------------------------------------------------------
#
df = pd.read_excel('./Ecliptic_ra-radian_dec-degree.xlsx', sheet_name='ecliptic')
#
ra_radian = []
dec_degrees = []
for i in range(0, len(df)) :
    ra_radian.append(  -1*df.at[i, 'ra_radian'])
    dec_degrees.append(df.at[i, 'dec_degrees'])
# end for
#
# function that draws each frame of the animation
def animate(i):
    ax.clear()
    #
    plt.xlim(-2*np.pi, np.pi)    # 設定X軸的顯示範圍, from xmin to xmax
    plt.ylim(-1.5, 1.5)          # 設定Y軸的顯示範圍, from ymin to ymax
    #
    xvalues = [-12*np.pi/6, -11*np.pi/6, -10*np.pi/6, -9*np.pi/6, -8*np.pi/6,
            -7*np.pi/6,  -6*np.pi/6,  -5*np.pi/6, -4*np.pi/6, -3*np.pi/6, 
            -2*np.pi/6,  -1*np.pi/6, 0]
    #
    yvalues = [-1.5,-1,-0.5,0,0.5,1,1.5]
    #
    plt.xticks(xvalues)
    plt.yticks(yvalues)
    plt.tick_params(labelsize=11)
    #
    ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
    #
    plt.xlabel("RA Hours",fontsize=12)
    plt.ylabel("Sin(Dec)",fontsize=12)
    #    
    # get current axes, set X,Y same ratio & scale 
    plt.gca().set_aspect('equal', adjustable='box')
    #
    plt.grid(color='greenyellow', linewidth=0.5)
    ax.grid(True)
    # ax.grid(False)
    ax.set_axis_on() # will turn on  grid line
    # ax.set_axis_off()  # will turn off grid line
    #
    circle_1 = shapely_Arc((np.pi/2,0), 1, 0, 360)
    motif_line(circle_1, 'blue', 10, 0.8)
    #
    RA_line = LineString([(-2*np.pi,0), (0,0)])
    motif_line(RA_line, 'red', 30, 0.6)
    #
    motif_line(Ecliptic_max, 'darkviolet', 30, 0.4)
    motif_line(Ecliptic_min, 'darkviolet', 30, 0.4)
    #
    ax.annotate("23.5°", text_1_xy,
            color='darkviolet',
            fontproperties=myfont,
            fontsize=9, 
            zorder=10)
    #
    ax.annotate("-23.5°", text_2_xy,
            color='darkviolet',
            fontproperties=myfont,
            fontsize=9, 
            zorder=10)
    #
    ax.set_title('Constellation Graph', fontsize=10, color='darkviolet')
    #
    motif_line(circle_1, 'blue', 10, 0.6)
    #
    # here enumerate(total_xs) should be 12
    for m, x_list in enumerate(total_xs) :
        ax.plot(total_xs[m], total_ys[m], color='royalblue', alpha=1.0, linewidth=0.7, linestyle='dotted', solid_capstyle='round', zorder=20)
        ax.fill(total_xs[m], total_ys[m], alpha=1.0, closed=True, edgecolor='none', facecolor=[228/255,235/255,243/255], fill=True, zorder=10)
    # end for
    #
    rod_x  = np.cos(ra_radian[i]) + np.pi/2
    rod_y  = np.sin(dec_degrees[i]*degree)        
    #
    rod_XS.append(rod_x)
    rod_YS.append(rod_y)
    #   
    rod_xs = [np.pi/2]
    rod_ys = [0]
    rod_xs.append(rod_x)
    rod_ys.append(rod_y)
    #
    ball_x = ra_radian[i]
    ball_y = rod_y
    ball = shapely_Arc((ball_x,ball_y), 0.05, 0, 360)
    motif_fill(ball, 'red', 30, 0.8)
    #    
    if i == 0 :
        j = 1
    else :
        sin_XS.append(ra_radian[i])
        sin_YS.append(rod_y)
    # end if
    #
    ax.plot(rod_xs,  rod_ys,  color='darkviolet', linewidth=0.6)
    ax.plot(rod_XS,  rod_YS,  color='chocolate',  linewidth=1.0)
    ax.plot(sin_XS,  sin_YS,  color='blue',  linewidth=0.8, zorder=30)
# end def
#
sin_XS = [0]
sin_YS = [0]
rod_XS = []
rod_YS = []
#
# run the animation
# frames=369 ; total running time is about 40 seconds
#
ani = animation.FuncAnimation(fig, animate, frames=366, interval=20, repeat=False)
#
# Saving Matplotlib animation in .gif format:
#
# f = r"./graphic/02-Constellation-animation-1.gif" 
# writergif = animation.PillowWriter(fps=30) 
# ani.save(f, writer=writergif)
#
```
