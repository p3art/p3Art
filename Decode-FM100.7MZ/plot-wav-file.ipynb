```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
#
plt.rcParams["figure.figsize"] = [29, 3.50]
plt.rcParams["figure.autolayout"] = True
input_data = read("./FM-100.7MZ-mono.wav")
audio = input_data[1]
#
# plt.plot(audio[0:1024])
plt.plot(audio)
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.show()
#
print('Done')
```
